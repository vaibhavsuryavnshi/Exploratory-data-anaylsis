```python
def check_skewness(df):
    skew = df.skew()
    for i,j in skew.items():
        if(j<-1)or (j>1):
            print(i,"is highly skewed")
        elif ((j > -1) or (j, -0.5)) or ((j> 0.5) or (j<1)):
            print(i,"is moderatelt skewed")
        else:
            print(i,"is normally distributed")
```

```python
check_skewness(df_sales)
```
